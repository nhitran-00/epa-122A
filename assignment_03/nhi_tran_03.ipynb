````python
# fill data for missing districts using the mean of their neighbors

def fill_missing_with_neighbor_mean(df, geo_code, neighbor_geo_codes):
    """
    fill missing values of a district using the mean of its neighbours
    ------------
    Parameters:
    df pd.DataFrame: the dataframe for a new row to be added
    geo_code str: the geo_code string of the district to be added to the input data frame
    neighbor_geo_codes list str: the geo_code strings of the neighbours of the district to be added to the input dataframe
    ------------
    Return:
    df pd.DataFrame: the dataframe with a new row
    """
    new_row = pd.Series({"geo_code":geo_code})
    mean_values = df[df.columns[1:]].loc[df["geo_code"].isin(neighbor_geo_codes)].mean(axis=0).apply(np.ceil)
    new_row = pd.concat([new_row,mean_values])
    df.loc[len(df.index)] = new_row
    return df


# add new rows to the df_socio_selection, 1 row = 1 missing district
for district in missing_districts:
    neighbor_geo_codes = df_geo_district.geo_code[df_geo_district.sindex.nearest(df_geo_district.geometry[df_geo_district.geo_code==district])[1]]
    df_socio_selection = fill_missing_with_neighbor_mean(df_socio_selection, district, neighbor_geo_codes)

````

### Spatial weight matrix

The steps to compute spatial weight matrix is as follows:
1. **create the contiguity queen weight matrix**
1. identify islands and disconnected components
1. **create the K-nearest-neighbor (KNN) weight matrix (k=2)**
1. **union the queen and the KNN weight matrix to handle islands**
1. inspect the disconnected components
2. drop disconnected components that might distort the analysis
3. create a final spatial weight matrix by repeating step **1, 3, 4**
4. row-standardise the spatial weight matrix

Using the contiguity queen rules to define neighbours, there are 25 disconnected components and 12 islands. So, another weight matrix is created following the KNN rules. Following the KNN rules, there are 3 disconnected components and 0 islands. As these 2 matrices are combined, there are 2 disconnected components and 0 islands.

Among the 2 components, one is mainland India, and the other one is a strip of islands in the sea. The strip of island is removed from the data to avoid distorting the analysis.

Local Indicators of Spatial Association (LISA) analysis shows that for districts that are significantly regarding total facility count and average household size, there is a negative correlation between the average household size and the total facility count. Therefore, building regression models for hypothesis analysis is not appropriate for this dataset. On the other hand, the clustering approach is more appropriate because it enables a more localised approach for analysis. In particular, rather than assuming the hypothesis holds true for the entire country, clustering can help identify areas where there are remarkable characteristics for further analyses.

## Build model - clustering

The steps taken are as follows:
1. Perform PCA
	1. standardise features
	1. Perform PCA
	1. Look at variance to identify components to select
1. Run clustering
    1. Use dendogram to identify the best number of clusters
    1. Run clustering model on the data

````python
# Standardise independent and dependent variables

std_vars_independent = []
std_vars_dependent = []

for col in vars_independent:
    new_col = "%s_std" % col
    std_vars_independent.append(new_col)
    geo_pop[new_col] = StandardScaler().fit_transform(np.array(geo_pop[col]).reshape(-1,1))

for col in vars_dependent:
    new_col = "%s_std" % col
    std_vars_dependent.append(new_col)
    geo_pop[new_col] = StandardScaler().fit_transform(np.array(geo_pop[col]).reshape(-1,1))
````

````python
# perform PCA on data and create a new dataframe from the PCA data
pca = PCA().fit(geo_pop[[*std_vars_independent, *std_vars_dependent]])
vars_pca = pca.transform(geo_pop[[*std_vars_independent, *std_vars_dependent]])
pca_df = pd.DataFrame(vars_pca, columns=["PCA%d" %i for i in range(1, vars_pca.shape[1] + 1)])
````

````python
# plot the variance ratio

f, ax = plt.subplots(1)

bars = ax.bar(pca_df.columns, pca.explained_variance_ratio_, align='center',
             label='individual explained variance')
ax.bar_label(bars, fmt='%.2f')

ax.step(pca_df.columns, np.cumsum(pca.explained_variance_ratio_), where='mid', 
         label='cumulative explained variance')

plt.ylim(0, 1.1)
plt.xlabel('Principal components')
plt.ylabel('Explained variance ratio')
plt.legend(loc='best')
plt.show()
````

The explained variance ratio shows that:
1. The first 3 components sum up to 99% of the relationship of all the variables.
1. The 4th component accounts for about 1%

Therefore, using the first 3 components is more than enough in clustering model in the next step. Using only the first 2 components, could impact the clustering results. However, since these 2 components make up for approximately 80% of the relationship of all the variables, the impact should be negligible.

````python
principals = ["PCA1", "PCA2", "PCA3"]
````

````python
np.random.seed(1234)
plt.figure(figsize =(8, 8))
plt.title('Dendogram of clusters')
Dendrogram = shc.dendrogram((shc.linkage(pca_df[principals], method ='ward')))
plt.axhline(y = 22, color = 'b', linestyle = '-', label="Number of clusters") 
plt.legend()
plt.show()
````
