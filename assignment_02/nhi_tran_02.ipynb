In India, there are three types of facilities run by the State to help Indian in family planning, namely Primary Health Sub Center, Maternity And Child Welfare Centre, and Family Welfare Centre.
````python
# create a copy of the population census data with only selected columns
socio_facility_vars = ["pc11_vd_phs_cntr", "pc11_vd_mcw_cntr", "pc11_vd_fwc_cntr"]
socio_population_vars = ["pc11_vd_t_p", "pc11_vd_t_hh"]

df_socio_selection = df_socio[["geo_code", *socio_facility_vars, *socio_population_vars]].copy()
df_socio_selection.head()
````

````python
# fill data for missing districts using the mean of their neighbors

def fill_missing_with_neighbor_mean(df, geo_code, neighbor_geo_codes):
    """
    fill missing values of a district using the mean of its neighbours
    ------------
    Parameters:
    df pd.DataFrame: the dataframe for a new row to be added
    geo_code str: the geo_code string of the district to be added to the input data frame
    neighbor_geo_codes list str: the geo_code strings of the neighbours of the district to be added to the input dataframe
    ------------
    Return:
    df pd.DataFrame: the dataframe with a new row
    """
    new_row = pd.Series({"geo_code":geo_code})
    mean_values = df[df.columns[1:]].loc[df["geo_code"].isin(neighbor_geo_codes)].mean(axis=0).apply(np.ceil)
    new_row = pd.concat([new_row,mean_values])
    df.loc[len(df.index)] = new_row
    return df


# add new rows to the df_socio_selection, 1 row = 1 missing district
for district in missing_districts:
    neighbor_geo_codes = df_geo_district.geo_code[df_geo_district.sindex.nearest(df_geo_district.geometry[df_geo_district.geo_code==district])[1]]
    df_socio_selection = fill_missing_with_neighbor_mean(df_socio_selection, district, neighbor_geo_codes)

````

### Spatial weight matrix

The steps to compute spatial weight matrix is as follows:
1. **create the contiguity queen weight matrix**
1. identify islands and disconnected components
1. **create the K-nearest-neighbor (KNN) weight matrix (k=2)**
1. **union the queen and the KNN weight matrix to handle islands**
1. inspect the disconnected components
2. drop disconnected components that might distort the analysis
3. create a final spatial weight matrix by repeating step **1, 3, 4**
4. row-standardise the spatial weight matrix

Using the contiguity queen rules to define neighbours, there are 25 disconnected components and 12 islands. So, another weight matrix is created following the KNN rules. Following the KNN rules, there are 3 disconnected components and 0 islands. As these 2 matrices are combined, there are 2 disconnected components and 0 islands.

Among the 2 components, one is mainland India, and the other one is a strip of islands in the sea. The strip of island is removed from the data to avoid distorting the analysis.

Inspecting the dropped districts, it can be seen that these districts:
1. have relatively small population and household count. Their **avg_hh_size** falls under the 1st quantile of the distribution.
1. have relatively low **facility**, which falls under the 1st quantile of the distribution.

These districts might fall into the population groups that are of particular interest to the analysis. Furthermore, since they are separated from the mainlaind of India, it is likely that they are often neglected by policy-makers in decision-making processes. Therefore, they might be the people that would benefit the most from additional attentions.

````python
# create spatial weight
w_queen = weights.Queen.from_dataframe(geo_pop, ids='geo_code')

# create the component label in the dataframe 
geo_pop.insert(len(geo_pop.columns), "comp_label", w_queen.component_labels)
geo_pop.comp_label = geo_pop.comp_label.astype("category")
````

````python
# create KNN weight
w_knn = weights.KNN.from_dataframe(geo_pop, ids='geo_code')
````

````python
# union queen and knn weights
w = weights.set_operations.w_union(w_queen, w_knn)

# update the component label column 
geo_pop.comp_label = w.component_labels
geo_pop.comp_label = geo_pop.comp_label.astype("category")
````

````python
# plot the component label to inspect the geo-spatial distributions of the 2 components
ax = geo_pop.plot(column = "comp_label", figsize=(9, 9), aspect="equal", legend=True, cmap="viridis")
````

````python
# calculate the Moran's I
mi_facility = esda.Moran(geo_pop['facility'], w)

print("Moran's I = %.3f" % mi_facility.I)
print("Moran's p-value = %.3f" % mi_facility.p_sim)

# plot the Moran's scatter plot
f, ax = plt.subplots(1, figsize=(9, 9))
moran_scatterplot(mi_facility, ax=ax)
ax.set_aspect("equal")
````

Moran's I = 0.205 and the p_value = 0.001 (0.1%) indicates that the facility spatial distribution isn't random.

````python
# calculate the LISA values
lisa_facility = esda.Moran_Local(geo_pop['facility'], w)

# Break observations into significant or not
geo_pop['significant_facility'] = lisa_facility.p_sim < 0.05
# Store the quadrant they belong to
geo_pop['quadrant_facility'] = lisa_facility.q


f, ax = plt.subplots(1, 3, figsize=(18,18))

# plot the choropleth of facility - all districts
geo_pop.plot(ax=ax[0], column='facility', 
            legend=True, aspect="equal",
            scheme='UserDefined', classification_kwds=bin_facility)
ax[0].set_axis_off()
ax[0].set_title("Facility distribution")

# plot the LISA clusters
lisa_cluster(lisa_facility, geo_pop, ax=ax[1])
ax[1].set_axis_off()
ax[1].set_aspect("equal")
ax[1].set_title("LISA quadrants")

# plot the choropleth of facility - statistically significant district
geo_pop.plot(ax=ax[2], linewidth=1, alpha=0.2, color="gray", aspect="equal")
geo_pop[geo_pop.significant_facility == True].plot(ax=ax[2], column="facility",
                                          legend=True, aspect="equal",
                                          scheme='UserDefined', classification_kwds=bin_facility)
ax[2].set_axis_off()
ax[2].set_aspect("equal")
ax[2].set_title("Facility distribution - LISA quadrants")

plt.show()
````
